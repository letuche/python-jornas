# Novo tipo de dados: dicionário (dict)
## Cada chave é associada a um valor
### - Chaves podem ser de (quase) qualquer tipo (nas listas são sempre números e não temos controle)
### - Valores podem ser de qualquer tipo (como nas listas)
## Exemplo:
### d = {"linguagem": "python", 42: "número"}
### d["linguagem"] devolve "python"
### d[42] devolve "número"

## Lendo arquivo csv como na aula anterior:

```python
import csv

arquivo = open('brasil.csv', encoding='utf8')
for linha in csv.reader(arquivo):
    print(linha)
```

## Lendo o arquivo como um dicionário:

```python
import csv

arquivo = open('brasil.csv', encoding='utf8')
for registro in csv.DictReader(arquivo):
    print(registro)
```

### O programa associa cada registro à respectiva coluna (chave), formando pares ('chave', 'valor') e eliminando o cabeçalho.

## Utilizando as chaves como índice para exibição dos registros:

```python
import csv

arquivo = open('brasil.csv', encoding='utf8')
for registro in csv.DictReader(arquivo):
    print(registro['municipio'])
```

## Repetindo o exercício da aula anterior, mas usando dicionário:

```python
import csv

print('\nCalculadora de população dos estados brasileiros')
estado_desejado = input('\nDigite a sigla do estado desejado: ')
estado_desejado = estado_desejado.upper()
populacao = 0
arquivo = open('brasil.csv', encoding='utf8')
for registro in csv.DictReader(arquivo):
    if registro['estado'] == estado_desejado:
        populacao += int(registro['habitantes'])
print('\nA população do estado ' + estado_desejado + ' é de: ' + str(populacao) + ' habitantes')
```

# String formatada
## Modelo que automaticamente converte as variáveis para string e as insere no texto predeterminado
### Em vez de 'modelo com ' + str(variavel) -> f'modelo com {variavel}'
### Exemplo: f'O valor é: {numero}' devolve 'O valor é: 42'

```python
import csv

print('\nCalculadora de população dos estados brasileiros')
estado_desejado = input('\nDigite a sigla do estado desejado: ')
estado_desejado = estado_desejado.upper()
populacao = 0
arquivo = open('brasil.csv', encoding='utf8')
for registro in csv.DictReader(arquivo):
    if registro['estado'] == estado_desejado:
        populacao += int(registro['habitantes'])
print(f'\nA população do estado {estado_desejado} é de {populacao} habitantes')
```

# Exercício:
## Receber a sigla de um estado. Para cada município desse estado, calcular e mostrar a densidade demográfica.
## Algoritmo:
### Abrir arquivo csv com municípios
### Para cada registro(município):
#### Se o estado for o estado digitado:
##### Calcula densidade (habitantes / area)
##### Mostra o nome do município e a densidade calculada

```python
import csv

estado_desejado = input('Digite a sigla do estado desejado: ')
estado_desejado = estado_desejado.upper()
arquivo = open('brasil.csv', encoding='utf8')
for registro in csv.DictReader(arquivo):
    if registro['estado'] == estado_desejado:
        densidade = int(registro['habitantes']) / float(registro['area'])
        print(f'Densidade demográfica do município de {registro["municipio"]} = {densidade:.1f} hab/km²')
```
